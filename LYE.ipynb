```python
from pandas_datareader import data
```

```python
data_FB=data.DataReader(name='FB', #ticker of staock
                        data_source='yahoo', #source of data, for example 'fred' is another data source
                        start='2020-01-01', end='2021-01-01', #time period
                        pause=0.1 # to puase some time in seconds after each request to Yahoo, in order not to get blocked
                        )
```

```python
data_FB.head()
```

```python
data_FB.Close.mean()
```

```python
data_FB.Close.median()
```

```python
fig=go.Figure()

fig.add_trace(go.Scatter(x=data_FB.index,
                        y=data_FB.Close,
                        mode='lines',
                        name='Close Price of FB'))

fig.add_trace(go.Scatter(x=data_FB.index,
                        y=data_FB.Open,
                        mode='lines',
                        name='Open Price of FB'))

fig.update_layout(title='Comparison of Close and Open prices of FB')

fig.show()
```
